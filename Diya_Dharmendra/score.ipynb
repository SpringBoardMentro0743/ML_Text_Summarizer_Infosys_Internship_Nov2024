```python
import pandas as pd

# Example dataset with input texts and reference summaries
data = {
    'input_texts': ["Input text 1", "Input text 2"],
    'reference_summaries': ["Reference summary 1", "Reference summary 2"]
}
df = pd.DataFrame(data)
```

```python
def generate_summary(input_text):
    input_ids = tokenizer.encode(input_text, return_tensors="pt", max_length=512, truncation=True)
    outputs = model.generate(input_ids, max_length=150, num_beams=4, early_stopping=True)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

df['generated_summaries'] = df['input_texts'].apply(generate_summary)
```

```python
from google.colab import drive
drive.mount('/content/drive')
!pip install transformers datasets sentencepiece
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Path to your saved model in Google Drive
model_path = "/content/drive/MyDrive/model2"

# Load the model and tokenizer
tokenizer = T5Tokenizer.from_pretrained(model_path)
model = T5ForConditionalGeneration.from_pretrained(model_path)
import pandas as pd

# Example dataset with input texts and reference summaries
data = {
    'input_texts': ["Input text 1", "Input text 2"],
    'reference_summaries': ["Reference summary 1", "Reference summary 2"]
}
df = pd.DataFrame(data)
def generate_summary(input_text):
    input_ids = tokenizer.encode(input_text, return_tensors="pt", max_length=512, truncation=True)
    outputs = model.generate(
        input_ids,
        max_length=150,
        min_length=50,
        num_beams=6,
        no_repeat_ngram_size=3,
        length_penalty=1.5,
        early_stopping=True,
        temperature=0.7,
        top_p=0.95
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)

df['generated_summaries'] = df['input_texts'].apply(generate_summary)
# Install or upgrade datasets and restart the runtime to make sure the changes take effect
!pip install datasets --upgrade
!pip install evaluate
!pip install rouge_score
import evaluate

# Load ROUGE and BLEU metrics
rouge = evaluate.load("rouge")
bleu = evaluate.load("bleu")

# Format references and predictions
references = [[ref] for ref in df['reference_summaries']]
predictions = df['generated_summaries'].tolist()

# Compute ROUGE
rouge_scores = rouge.compute(predictions=predictions, references=references)
print("ROUGE Scores:", rouge_scores)

# Compute BLEU
# Removed the .split() from predictions and references
bleu_scores = bleu.compute(predictions=predictions, references=references)
print("BLEU Score:", bleu_scores)
```
